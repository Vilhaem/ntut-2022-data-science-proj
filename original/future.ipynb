```python
import pandas as pd

df = pd.read_csv('經濟部能源局_未來電力供需預測.csv')
print(df.columns)
df.head()
```

```python
df.isnull().sum()
```

```python
df['負載預測(萬瓩)'] = df['負載預測(萬瓩)'] * 10
df['淨尖峰供電能力(萬瓩)'] = df['淨尖峰供電能力(萬瓩)'] * 10

df.head()
```

```python
new_df = df.copy()

new_df.rename(columns = {
    '年度': 'year',
    '負載預測(萬瓩)': 'load forecast(MW)',
    '淨尖峰供電能力(萬瓩)': 'net peak power(MW)',
    '備用容量率(%)': 'reserve margin rate(%)',
    '備轉容量率(%)': 'operating reserve rate(%)'
}, inplace = True)

new_df.head()
```

```python
new_df.to_csv('future_predict.csv', encoding='utf-8', index=False)
```
