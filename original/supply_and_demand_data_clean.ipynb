```python
import pandas as pd

df = pd.read_csv('origin_supply_and_demand.csv')
df.columns
```

```python
new_df = df[['日期(年/月)', '總發電量',  '總發電量_發電業', '總發電量_發電業_抽蓄水力', '總發電量_發電業_火力', '總發電量_發電業_核能', '總發電量_發電業_再生能源', '總發電量_自用發電設備', '總需要量']].copy()

new_df.rename(columns = {
    '日期(年/月)':'year-month(date)', 
    '總發電量':'total power generation',
    '總發電量_發電業': 'power supply',
    '總發電量_發電業_抽蓄水力': 'hydroelectricity',
    '總發電量_發電業_火力': 'thermal power',
    '總發電量_發電業_核能': 'nuclear',
    '總發電量_發電業_再生能源': 'Renewable energy',
    '總發電量_自用發電設備': 'self use',
    '總需要量': 'total power requirement'
}, inplace = True)

new_df['year'] = new_df['year-month(date)'].apply(lambda date: int(date / 100))
new_df['month'] = new_df['year-month(date)'].apply(lambda date: int(date % 100))
new_df = new_df.drop(['year-month(date)'], axis = 1)
new_df
```

```python
new_df.to_csv('supply_and_demand.csv', encoding='utf-8', index=False)
```
