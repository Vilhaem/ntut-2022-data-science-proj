```python
import pandas as pd

df = pd.read_csv("台灣電力公司_各縣市年度住宅、服務業、機關及工業用電比例.csv")
print(df.columns)
df.head()
```

```python
df.isnull().sum()
```

```python
new_df = df[['年度', '縣市', 'A.住宅部門', 'B.服務業部門', 'C.機關用電(政府+包燈+大專院校)', 'D.工業', 'E.其他']].copy()

new_df.rename(columns = {
    '年度': 'year',
    '縣市': 'county',
    'A.住宅部門': 'residential',
    'B.服務業部門': 'service industry',
    'C.機關用電(政府+包燈+大專院校)': 'public department',
    'D.工業': 'industry',
    'E.其他': 'other'
}, inplace = True)

new_df.head()
```

```python
new_df.to_csv('power_ratio.csv', encoding='utf-8', index=False)
```
