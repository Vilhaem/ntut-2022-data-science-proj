```python
import pandas as pd

df = pd.read_csv('用電統計資料.csv')
print(df.columns)
df.head()
```

```python
df.isnull().sum()
```

```python
new_df = df.copy()

new_df.rename(columns = {
    '年度': 'year',
    ' 電燈售電量(度) ': 'electric lights(kWh)',
    ' 電力售電量(度) ': 'electricity sales(kWh)',
    ' 售電量合計(度) ': 'total amount(kWh)',
    ' 電燈用戶數(戶) ': 'electric lights(households)',
    ' 電力用戶數(戶) ': 'electricity sales(households)',
    ' 用戶數合計(戶) ': 'total households',
    ' 電燈(非營業用)用戶數(戶) ': 'electric lights households(non-commercial)',
    ' 電燈(營業用)用戶數(戶) ': 'electric lights households(commercial)',
    '電燈平均電價(元)': 'average price of electric lights',
    '電力平均電價(元)': 'average price of electricity sales',
    '平均電價合計(元)\t\t\t\t\t\t\t\t\t\t\t': 'average price of total amount'
}, inplace = True)

new_df.head()
```

```python
new_df.to_csv('power_consumption.csv', encoding='utf-8', index=False)
```
